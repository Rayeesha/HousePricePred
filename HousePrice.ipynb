Importing dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
ds=fetch_california_housing()
```

```python
df=pd.DataFrame(ds.data,columns=ds.feature_names)
```

```python
df.head()
```

```python
df['Price']=ds.target
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.describe()
```

Positive correlation - Directly Proportional
Negative Correlation -Inversely Proportional

```python
cr = df.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(cr,annot=True,cbar=True,square=True,fmt='.1f',annot_kws={'size':8},cmap='Blues')
```

Separating  data and labels

```python
X=df.drop(columns='Price',axis=1)
Y=df["Price"]
```

```python
print(X,Y)
```

Splitting it as train and test

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=3)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training

```python
model=XGBRegressor()
```

```python
model.fit(X_train,Y_train)
```

Evaluation

Prediction on training data

```python
train_pred=model.predict(X_train)
print(train_pred)
```

```python
#R squared error
error_score=metrics.r2_score(Y_train,train_pred)
print("R squared error : ",error_score)
```

```python
error_score2=metrics.mean_absolute_error(Y_train,train_pred)
print("Mean absolute error:",error_score2)
```

Predicting on testing data

```python
test_pred=model.predict(X_test)
print(test_pred)
```

```python
#R squared error
error_score=metrics.r2_score(Y_test,test_pred)
print("R squared error : ",error_score)
error_score2=metrics.mean_absolute_error(Y_test,test_pred)
print("Mean absolute error:",error_score2)
```

Visualizing through Scatter Plot

```python
plt.scatter(Y_train,train_pred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Price vs Predicted Price")
plt.show()
```
